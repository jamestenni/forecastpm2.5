## Prepare Train Data

```python
def read_data_hotspot():
    # read hotspot data
    df_hotspot = pd.read_csv('./hotspot_scraped_data_interpolated.csv')
    df_hotspot['date'] = pd.to_datetime(df_hotspot['date'], format = '%Y-%m-%d')   
    
    return df_hotspot

df_hotspot = read_data_hotspot()

def read_data_province(province):
    train_data_path = f'./datasci_dataset_2022/{province}/train'
    
    # read PM2.5 train data
    pm25_train = pd.read_csv(train_data_path + f'/{province.lower()}_train.csv')
    pm25_train.rename(columns = {pm25_train.columns[0]: 'date_time'}, inplace = True)
    pm25_train['date_time'] = pd.to_datetime(pm25_train['date_time'], format = '%Y-%m-%d %H:%M:%S')
    pm25_train.sort_values(by = 'date_time', inplace = True)

    # read wind train data
    wind_train = pd.read_csv(train_data_path + f'/{province.lower()}_weather_wind.csv')
    if (province in ["BKK", "Khonkaen", "Rayong"]):
        wind_train['date_time'] = pd.to_datetime(wind_train['date_time'], format = '%Y-%m-%d %H:%M:%S')
    elif (province in ["Chiangmai", "Surat"]):
        wind_train['date_time'] = pd.to_datetime(wind_train['date_time'], format = '%m/%d/%Y %H:%M')
    elif (province in ["Saraburi"]):
        wind_train['date_time'] = pd.to_datetime(wind_train['date_time'], format = '%d/%m/%Y %H:%M')
    
    wind_train.rename(columns = {'wind speed': 'wind_speed', 'wind dir': 'wind_dir'}, inplace = True)
    wind_train.sort_values(by = 'date_time', inplace = True)

    # read temp train data
    temp_train = pd.read_csv(train_data_path + f'/{province.lower()}_temp_surface.csv')
    
    if (province in ["BKK", "Chiangmai", "Rayong", "Saraburi", "Surat"]):
        temp_train['date_time'] = pd.to_datetime(temp_train['date_time'], format = '%m/%d/%Y %H:%M')
    elif (province in ["Khonkaen"]):
        temp_train['date_time'] = pd.to_datetime(temp_train['date_time'], format = '%Y-%m-%d %H:%M:%S')
    temp_train.sort_values(by = 'date_time', inplace = True)
    
    return pm25_train, wind_train, temp_train
```

```python
DAY_HOTSPOT_COLUMNS = df_hotspot.columns[[i for i in range (1, len(df_hotspot.columns), 2)]]
NIGHT_HOTSPOT_COLUMNS = df_hotspot.columns[[i for i in range (2, len(df_hotspot.columns), 2)]]

REGIONS = ['Thailand', 'Myanmar', 'Cambodia', 'Lao_PDR', 'P_Malaysia'] #P_Malaysia = Peninsula Malaysia

def get_hotspot_column(date_time_series, df_hotspot):
    hotspot_DATA = []
    column_names = ['date_time'] + [f'hotspot_{r.lower()}' for r in REGIONS]  
    
    for dt in date_time_series:
        date = dt.date()
        hour = dt.time().hour
        
        if (hour >= 6 and hour < 18):
            daytime = 'day'
        else:
            daytime = 'night'
            
        data = pd.Series(data = [dt], index = ['date_time'])
        if (daytime == 'day'):
            hotspot_data = df_hotspot.loc[df_hotspot['date'] == pd.to_datetime(str(date), format = '%Y-%m-%d'), DAY_HOTSPOT_COLUMNS].iloc[0] / 12
        elif (daytime == 'night'):
            hotspot_data = df_hotspot.loc[df_hotspot['date'] == pd.to_datetime(str(date), format = '%Y-%m-%d'), NIGHT_HOTSPOT_COLUMNS].iloc[0] / 12  
        hotspot_data = np.ceil(hotspot_data)
        data = pd.concat([data, hotspot_data])
        data.index = column_names               
        hotspot_DATA.append(data)
        
    return pd.DataFrame(hotspot_DATA)


def combined_data(pm25_train, wind_train, temp_train, df_hotspot):
    # combine pm25, wind, temp
    train = pd.merge(pd.merge(pm25_train, wind_train, on = 'date_time', how = 'outer'), temp_train, on = 'date_time', how = 'outer')
    
    # combine hotspot
    hotspot_cols = get_hotspot_column(train['date_time'], df_hotspot)
    train = pd.merge(train, hotspot_cols, on = 'date_time', how = 'outer')
    
    return train


def perform_interpolation(df_combined):
    df = df_combined.copy()
    
    # perform linear interpolation (PM2.5, wind_speed, temp)
    df['PM2.5'] = df['PM2.5'].interpolate().round(0)
    df['wind_speed'] = df['wind_speed'].interpolate().round(0)
    df['temp'] = df['temp'].interpolate().round(1)
    
    # perform cyclic interpolation (wind_speed)
    df2 = df.loc[df['wind_dir'].notna(), ['date_time','wind_dir']].copy()
    df2.rename(columns = {'wind_dir': 'wind_dir_2'}, inplace = True)
    df2['wind_dir_2'] = np.rad2deg(np.unwrap(np.deg2rad(df2['wind_dir_2'])))
    
    df3 = pd.merge(df, df2, on = 'date_time', how = 'left')
    df3['wind_dir_2'] = df3['wind_dir_2'].interpolate().round(0)
    df3['wind_dir_2'] %= 360
    df3.drop(columns = 'wind_dir', inplace = True)
    df3.rename(columns = {'wind_dir_2': 'wind_dir'}, inplace = True)
    
    #rearange columns
    df3 = df3[df.columns].copy()
    
    return df3   
```

```python
TRAIN_TS = {'BKK': {'start': '2017-07-01 00:00:00', 'end': '2020-06-30 23:00:00'},
            'Chiangmai': {'start': '2017-07-01 00:00:00', 'end': '2020-06-30 23:00:00'},
            'Khonkaen': {'start': '2017-07-01 00:00:00', 'end': '2020-06-30 23:00:00'},
            'Rayong': {'start': '2017-07-01 00:00:00', 'end': '2020-06-30 23:00:00'},
            'Saraburi': {'start': '2017-07-01 00:00:00', 'end': '2020-06-30 23:00:00'},
            'Surat': {'start': '2018-10-01 00:00:00', 'end': '2020-06-30 23:00:00'} #different is here
            }

def check_missing_timestep(timestamps, freq, province, mode):    
    if (mode == "train"):
        start_ts = pd.to_datetime(TRAIN_TS[province]['start'], format = '%Y-%m-%d %H:%M:%S')
        end_ts = pd.to_datetime(TRAIN_TS[province]['end'], format = '%Y-%m-%d %H:%M:%S')
    elif (mode == "test"):
        start_ts = pd.to_datetime('2020-07-01 00:00:00', format = '%Y-%m-%d %H:%M:%S')
        end_ts = pd.to_datetime('2021-07-01 23:00:00', format = '%Y-%m-%d %H:%M:%S')
    
    complete_timestamps = pd.date_range(start = start_ts, 
                                        end = end_ts,
                                        freq = freq)
    return complete_timestamps.difference(timestamps)
```

```python
def get_training_data_single_province(province):
    pm25_train, wind_train, temp_train = read_data_province(province)
    train = combined_data(pm25_train, wind_train, temp_train, df_hotspot)
    train_interpolated = perform_interpolation(train)
    train_interpolated.sort_values(by = 'date_time', inplace = True)
    
    # filter interested time only!
    train_interpolated = train_interpolated[(train_interpolated['date_time'] >= TRAIN_TS[province]['start']) &
                                           (train_interpolated['date_time'] <= TRAIN_TS[province]['end'])]
    
    missing_timesteps = check_missing_timestep(timestamps = train_interpolated['date_time'],
                      freq = 'H', province = province, mode = 'train')

    return train_interpolated, missing_timesteps

def save_data(df, province, mode):
    '''
        mode - 'train' or 'test'
    '''
    # create folder if not exist yet
    path = f'./processed_data/{province}'
    if not os.path.exists(path):
        os.makedirs(path)
        
    df.to_csv(path + f'/{province.lower()}_{mode}.csv', index = False)
```

```python
PROVINCES = ['BKK', 'Chiangmai', 'Khonkaen', 'Rayong', 'Saraburi', 'Surat']

TRAIN_DATA = {}
print("Starting...")
for i in range (len(PROVINCES)):
    province = PROVINCES[i]
    
    print(f'Processing (Training) Province {i+1}/{len(PROVINCES)}: {province} ...')
    train_data, missing_timesteps = get_training_data_single_province(province)
    TRAIN_DATA[province] = train_data
    
    if len(missing_timesteps) != 0:
        print(f'There are {len(missing_timesteps)} missing timesteps!, which are ..')
        print(missing_timesteps)
    else:
        print('OK There is no missing timesteps :)')
    print()
print("Finish...")
```

```python
#saving train data
for province in PROVINCES:
    save_data(TRAIN_DATA[province], province, 'train')
```

## Prepare Test Data

```python
wind_and_temp_test_df = pd.read_csv('./wind_and_temp_combined_data.csv')
wind_and_temp_test_df['date_time'] = pd.to_datetime(wind_and_temp_test_df['date_time'], format = '%Y-%m-%d %H:%M:%S')

wind_and_temp_test_df
```

```python
def read_PM25_province_test(province):
    test_data_path = f'./datasci_dataset_2022/{province}/test'
    
    # read PM2.5 test data
    pm25_test = pd.read_csv(test_data_path + f'/{province.lower()}_test.csv')
    pm25_test.rename(columns = {pm25_test.columns[0]: 'date_time'}, inplace = True)
    pm25_test['date_time'] = pd.to_datetime(pm25_test['date_time'], format = '%Y-%m-%d %H:%M:%S')
 
    return pm25_test

def combined_test_data(pm25_test, wind_and_temp_test_df, df_hotspot, province):
    df = pm25_test.copy()
    
    # combine with wind and temp scraped
    selected_col = ['date_time', f'wind_speed_{province}', f'wind_direction_{province}', f'temp_{province}']
    df_wt = wind_and_temp_test_df[selected_col].copy()
    df_wt.rename(columns = {f'wind_speed_{province}': 'wind_speed',
                            f'wind_direction_{province}': 'wind_dir',
                            f'temp_{province}': 'temp'}, inplace = True)
    
    df = pd.merge(df, df_wt, on = 'date_time', how = 'left')
    
    # combine with hotspot
    hotspot_cols = get_hotspot_column(df['date_time'], df_hotspot)
    df = pd.merge(df, hotspot_cols, on = 'date_time', how = 'outer')
    
    return df
```

```python
def get_testing_data_single_province(province):
    pm25_test = read_PM25_province_test(province)
    test = combined_test_data(pm25_test, wind_and_temp_test_df, df_hotspot, province)
    test_interpolated = perform_interpolation(test)
    test_interpolated.sort_values(by = 'date_time', inplace = True)
    
    # fill the missing value in first row by copy value of the second row
    test_interpolated.loc[0, 'wind_dir'] = test_interpolated.loc[1, 'wind_dir']
    test_interpolated.loc[0, 'wind_speed'] = test_interpolated.loc[1, 'wind_speed']
    test_interpolated.loc[0, 'temp'] = test_interpolated.loc[1, 'temp']
    
    missing_timesteps = check_missing_timestep(timestamps = test_interpolated['date_time'],
                      freq = 'H', province = province, mode = 'test')

    return test_interpolated, missing_timesteps
```

```python
PROVINCES = ['BKK', 'Chiangmai', 'Khonkaen', 'Rayong', 'Saraburi', 'Surat']

TEST_DATA = {}
print("Starting...")
for i in range (len(PROVINCES)):
    province = PROVINCES[i]
    
    print(f'Processing (Testing Data) Province {i+1}/{len(PROVINCES)}: {province} ...')
    test_data, missing_timesteps = get_testing_data_single_province(province)
    TEST_DATA[province] = test_data
    
    if len(missing_timesteps) != 0:
        print(f'There are {len(missing_timesteps)} missing timesteps!, which are ..')
        print(missing_timesteps)
    else:
        print('OK There is no missing timesteps :)')
    print()
print("Finish...")
```

```python
#saving test data
for province in PROVINCES:
    save_data(TEST_DATA[province], province, 'test')
```
